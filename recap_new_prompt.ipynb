```python
# Evaluation
from bert_score import score

def calculate_bert_score(generated_defs, reference_defs):
    # Calculate BERTScore
    P, R, F1 = score(generated_defs, reference_defs, model_type='microsoft/deberta-v3-large')

    # P: Precision scores
    # R: Recall scores
    # F1: F1 scores

    return P, R, F1
```

```python
bert_score = calculate_bert_score(gemini_df['Definition'].to_list(), reference_df['definition'].to_list())
```

```python
precision, recall, F1 = bert_score
```

```python
import numpy as np
import torch

def compute_mean_std(precision, recall, f1):
  precision_mean = torch.mean(precision)
  precision_std = torch.std(precision)

  recall_mean = torch.mean(recall)
  recall_std = torch.std(recall)

  f1_mean = torch.mean(f1)
  f1_std = torch.std(f1)
  pre = {
      'mean': (round(float(precision_mean), 5)) * 100,
      'std': (round(float(precision_std), 5)) * 100
  }
  rec = {
      'mean': (round(float(recall_mean), 5)) * 100,
      'std': (round(float(recall_std), 5)) * 100
  }
  f1 = {
      'mean': (round(float(f1_mean), 5)) * 100,
      'std': (round(float(f1_std), 5)) * 100
  }

  return pre, rec, f1

precision_mean = torch.mean(precision)
recall_mean = torch.mean(recall)
f1_mean = torch.mean(F1)
gemini_pre, gemini_rec, gemini_f1 = compute_mean_std(precision, recall, F1)
# print(f'precision_mean: {round(float(precision_mean), 5) * 100}')
print(f'precision_mean: {gemini_pre}')
# print(f'recall_mean: {round(float(recall_mean), 5) * 100}')
print(f'recall_mean: {gemini_rec}')
# print(f'F1_mean: {round(float(f1_mean), 5) * 100}')
print(f'F1_mean: {gemini_f1}')
```
