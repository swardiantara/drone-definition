```python
import os

print(os.path.abspath(os.curdir))
os.chdir("..")
print(os.path.abspath(os.curdir))
```

```python
# load the definition
definition_file = open(os.path.join('experiments', 'copilot', 'web-interface.txt'), "r")
definition_file.seek(0)
# readlines function
definition_dict = {
    'Term': [],
    'Definition': [],
    'Confidence': []
}
for line in definition_file.readlines():
    if line != '\n':
        if len(line.split(':')) > 1:
            key, value = line.split(':')
        else:
            key, value = 'Term', line
        definition_dict[key].append(value.strip())
definition_file.close()
```

```python
import pandas as pd

data = pd.DataFrame.from_dict(definition_dict)
data
```

```python
data.to_excel(os.path.join('experiments', 'copilot', 'web-interface.xlsx'))
```

```python
# load the definition from old-copilot.txt
definition_file = open(os.path.join('experiments', 'copilot', 'old-copilot.txt'), "r")
definition_file.seek(0)
# readlines function
definition_dict = {
    'term': [],
    'pred_definition': [],
}
curr_term = ''
curr_definition = ""
for line in definition_file.readlines():
    print(f'current line: {line}')
    if line != '\n':
        line_split = line.split(':')
        print(line_split)
        if len(line_split) > 1:
            if line_split[0] == 'Term' and curr_term == '':
                curr_term = line_split[1]
            elif line_split[0] == 'Term' and curr_term:
                definition_dict['term'].append(curr_term.strip())
                definition_dict['pred_definition'].append(curr_definition.strip())
                curr_term = line_split[1]
                curr_definition = ''
            elif line_split[0] == 'Definition' and curr_term and curr_definition == '':
                curr_definition = line_split[1]
            else:
                curr_definition = curr_definition + '\n' + line
        else:
            curr_definition = curr_definition + '\n' + line
    else:
        curr_definition += '\n'
    
if curr_term:
    definition_dict['term'].append(curr_term.strip())
    definition_dict['pred_definition'].append(curr_definition.strip())
definition_file.close()
```
