```python
# load the definition
definition_file = open(os.path.join('experiments', 'gemini', '2025-02-05_1738748338.txt'), "r")
definition_file.seek(0)
# readlines function
definition_dict = {
    'Term': [],
    'Definition': [],
    'Confidence': []
}
for line in definition_file.readlines():
    if line != '\n':
        key, value = line.split(':')
        definition_dict[key].append(value.strip())
definition_file.close()
```

```python
import pandas as pd

data = pd.DataFrame.from_dict(definition_dict)
data
```

```python
conf2idx = {
    'High': 3,
    'Medium': 2,
    'Low': 1
}
data['Conf'] = data['Confidence'].map(conf2idx)
```

```python
print(round(data['Conf'].mean(), 3))
```

```python
data.to_excel(os.path.join('experiments', 'gemini', '2025-02-05_1738748338.xlsx'))
```

```python
# Evaluation
from bert_score import score

def calculate_bert_score(generated_defs, reference_defs):
    # Calculate BERTScore
    P, R, F1 = score(generated_defs, reference_defs, model_type='microsoft/deberta-v3-base')
    
    # P: Precision scores
    # R: Recall scores
    # F1: F1 scores
    
    return P, R, F1
```
