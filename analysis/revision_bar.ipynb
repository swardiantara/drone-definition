```python
recap_dict = {
    'model': [],
    'precision_mean': [],
    'precision_std': [],
    'recall_mean': [],
    'recall_std': [],
    'f1_mean': [],
    'f1_std': [],
    'sum_distance': []
}
with open(os.path.join('..', 'recap_old.json')) as json_file:
    data = json.load(json_file)
    for model, metrics in data.items():
        recap_dict['model'].append(model)
        for key, value in metrics.items():
            recap_dict[key].append(value)
```
