```python
import os

print(os.path.abspath(os.curdir))
os.chdir("..")
print(os.path.abspath(os.curdir))
```

```python
# load the definition
definition_file = open(os.path.join('experiments', 'meta-ai', 'web-interface.txt'), "r")
definition_file.seek(0)
# readlines function
definition_dict = {
    'Term': [],
    'Definition': [],
    'Confidence': []
}
for line in definition_file.readlines():
    if line != '\n':
        key, value = line.split(':')
        definition_dict[key].append(value.strip())
definition_file.close()
```

```python
import pandas as pd

data = pd.DataFrame.from_dict(definition_dict)
data
```

```python
data.to_excel(os.path.join('experiments', 'meta-ai', 'web-interface.xlsx'))
```
